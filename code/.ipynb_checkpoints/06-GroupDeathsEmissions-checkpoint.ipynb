# Grouping files Deaths-Emissions by Pollutants

```python
import pandas as pd
import numpy  as np
import csv
from glob import glob
```

```python
FEmissions = pd.ExcelFile('../data/csv/Emissions/geo/Comp_by_Emi_total_Geo_2001-2015.xlsx', 
                       sheetname='Comp_by_Emi_total_Geo_2001-2015')
DfEmissions = pd.read_excel(FEmissions,index_col=None,na_values=['NA'])
DfEmissions = DfEmissions[(DfEmissions['MedioReceptor'] == 'Aire')]
```

```python
#Listado Ãºnico de emisiones
ListEmissions = DfEmissions["Contaminante"].unique()
del DfEmissions
```

```python
#Listado de ficheros entrada DeathsEmissions
ListFichDeathsEmissionsFinal = glob('../data/csv/merge/DeathsEmissionsFinal_*.csv')
```

```python
#Recorremos listado de emisiones y generamos fichero para cada uno de ellos con todos los registros 
#de los ficheros DeathsEmissions 

for i in range(0,len(ListEmissions)): 
    Df1  = pd.DataFrame()
    cont = 0
    print ('emision.:',i)
    
    for i2 in range(0,len(ListFichDeathsEmissionsFinal)):   
        
         print ('defu.:',i2)
            
         DfDeathsEmissions = pd.read_csv(ListFichDeathsEmissionsFinal[i2],sep=';',compression='gzip') #cargamos df con cada fichero
        
         Df1 = Df1.append(DfDeathsEmissions[DfDeathsEmissions['Contaminante'] == ListEmissions[i]])
            
         print ('df1.:',len(Df1))
 
         if len(Df1) > 1500000:   
            NomFich = ListEmissions[i].replace('(', '')
            NomFich = NomFich.replace(')', '')
            NomFich = NomFich.replace('/', '-')
            NomFich = NomFich.replace(' ', '_')
            NomFich = NomFich.replace(',', '_')
            NomFich = NomFich.replace('Ã¡', 'a')
            NomFich = NomFich.replace('Ã­', 'i')
            NomFich = NomFich.replace('Ã³', 'o')
            NomFich = NomFich.replace('Ã', 'O')
            NomFich = NomFich.replace('Contaminante_', 'C_')

            print(NomFich) 
            
            cont=cont+1
            Df1.to_csv('../data/csv/group/DeathsEmissions_'+str(NomFich)+'_'+str(cont)+'.csv', \
                       sep=';',compression='gzip')
            del Df1    
            Df1 = pd.DataFrame()
            
            
    if len(Df1) > 0:   
        NomFich = ListEmissions[i].replace('(', '')
        NomFich = NomFich.replace(')', '')
        NomFich = NomFich.replace('/', '-')
        NomFich = NomFich.replace(' ', '_')
        NomFich = NomFich.replace(',', '_')
        NomFich = NomFich.replace('Ã¡', 'a')
        NomFich = NomFich.replace('Ã­', 'i')
        NomFich = NomFich.replace('Ã³', 'o')
        NomFich = NomFich.replace('Ã', 'O')
        NomFich = NomFich.replace('Contaminante_', 'C_')

        print(NomFich)  
        
        cont=cont+1
        Df1.to_csv('../data/csv/group/DeathsEmissions_'+str(NomFich)+'_'+str(cont)+'.csv', \
                   sep=';',compression='gzip')
        del Df1    
        del DfDeathsEmissions
```
